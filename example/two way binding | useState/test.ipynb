```python
import nmap

nm = nmap.PortScanner()

# Replace '127.0.0.1' with the IP address you want to scan
nm.scan('27.79.181.232', '22-443')

vulnerabilities = []

for host in nm.all_hosts():
    for proto in nm[host].all_protocols():
        lport = nm[host][proto].keys()
        for port in lport:
            if nm[host][proto][port]['state'] == 'open':
                vulnerabilities.append((host, port))

for v in vulnerabilities:
    print(f"Potential vulnerability: Host {v[0]} has open port {v[1]}")
```

```python
from scapy.all import ARP, Ether, srp

ip_address = "27.79.181.232/24"  # replace with the IP address you want to scan

# Create an ARP request packet
arp = ARP(pdst=ip_address)

# Create an Ethernet frame
ether = Ether(dst="ff:ff:ff:ff:ff:ff")

# Combine them
packet = ether/arp

result = srp(packet, timeout=3, verbose=0)[0]

# Parse the response
for sent, received in result:
    print(f"{received.psrc} is online")
```
